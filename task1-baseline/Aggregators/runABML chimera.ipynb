```python
import subprocess

# Define the static variables
data_source = r"/data/pa_cpgarchive/projects/chimera/bladder/pathology_data/processed/features_erasmus_task2_uni/features_erasmus_task2_uni/features"

batch_size = 4
results_dir = "/data/temporary/nadieh/chimera/results"
split_names = "train,test"
in_dim = 1024
epoch = 13
fold = 5
task = 'mutation'
model_configs = [ "ABMIL_default"]
label = 'label'
# Loop over k from 1 to 5
for model_config in model_configs:
    model_type = "ABMIL"
    for k in range(0, fold):
        split_dir = f"/data/temporary/nadieh/PANTHER/splits_chimera/fold_{k}/"
        # Build the command
        cmd = [
            'python3', 'training/main_classification.py',
            '--data_source', data_source,
            '--split_dir', split_dir,
            '--task', task,
            '--batch_size', str(batch_size),
            '--results_dir', results_dir,
            '--split_names', split_names,
            '--in_dim', str(in_dim),
            '--model_config', model_config,
            '--target_col', label,
            '--model_type', model_type,
            '--max_epochs', str(epoch),
            

        ]
        # Execute the command
        subprocess.run(cmd)
```
