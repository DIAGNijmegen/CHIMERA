```python
feat='extracted-vit_large_patch16_224.dinov2.uni_mass100k'
input_dim=1024
mag='20x'
patch_size=256
model = 'OT'

bag_size='-1'
batch_size=64
out_size=16
out_type='allcat'
model_tuple='OT_default'
lin_emb_model='LinearEmb'
max_epoch=50
lr=0.0001
wd=0.00001
lr_scheduler='cosine'
opt='adamW'
grad_accum=1
loss_fn='cox'
n_label_bin=4
alpha=0.5
em_step=1
load_proto=1
es_flag=0
tau=1.0
eps=1
n_fc_layer=0
proto_num_samples='1.0e+06'
save_dir_root='result'




task='BRCA_survival'
target_col='dss_survival_days'
split_names='train,test'

model_config = f'{model}_default'
feat_name = feat.replace('extracted-', '')
exp_code = f'{task}::{model_config}::{feat_name}'
save_dir = f'{save_dir_root}/{exp_code}'

th = 0.00005
warmup = 0 if lr <= th else 1
curr_lr_scheduler = 'constant' if warmup == 0 else lr_scheduler

# Path for feature directories
all_feat_dirs = "/data/temporary/nadieh/patches_uni/features/tcga_brca/extracted_mag20x_patch256_fp/extracted-vit_large_patch16_224.dinov2.uni_mass100k/feats_pt"

# Loop through different splits
for i in range(0,5):  # Loop from 0 to 4
    split_dir = f'/data/temporary/nadieh/PANTHER/src/splits/classification/classificationBRCA/TCGA_BRCA_overall_survival_k={i}/'
    
    # Build the command
    cmd = [
        'python3','training/main_survival.py',
        '--data_source', all_feat_dirs,
        '--results_dir', save_dir,
        '--split_dir', os.path.join(split_dir),
        '--split_names', split_names,
        '--task', task,
        '--target_col', target_col,
        '--model_type', model,
        '--model_config', f'{model}_default',
        '--in_dim', str(input_dim),
        '--opt', opt,
        '--lr', str(lr),
        '--lr_scheduler', curr_lr_scheduler,
        '--accum_steps', str(grad_accum),
        '--wd', str(wd),
        '--warmup_epochs', str(warmup),
        '--max_epochs', str(max_epoch),
        '--train_bag_size', bag_size,
        '--batch_size', str(batch_size),
        '--in_dropout', '0',
        '--seed', '1',
        '--num_workers', '6',
        '--em_iter', str(em_step),
        '--tau', str(tau),
        '--n_proto', str(out_size),
        '--out_type', out_type,
        '--emb_model_type', lin_emb_model,
        '--ot_eps', str(eps),
        '--fix_proto',
        '--loss_fn', loss_fn,
        '--model_config', model_tuple
    ]
# Add prototype path if load_proto is True
    if load_proto == 1:
        proto_path = os.path.join(split_dir, f'prototypes/prototypes_c{out_size}_extracted-{feat_name}_faiss_num_{proto_num_samples}.pkl')
        cmd.extend(['--load_proto', '--proto_path', proto_path])

# Execute the command
    subprocess.run(cmd)
```
